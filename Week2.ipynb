````
# GET IMG, SHOW N PLOT SIZE

import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/kitten01.jpg')
plt.imshow(img)
````

````
# CONVERT 2 RGB
img2 = img
img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img3)
````

````
# GRAYSCALE IMG
img_gray = cv.imread('/content/drive/MyDrive/kitten01.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
````

````
# GRAY IMG
plt.imshow(img_gray, cmap='gray')
````

````
# MAGMA IMG
plt.imshow(img_gray, cmap='magma')
````

````
# RGB N RESIZE IMG 512 x 1024
img4 = cv.resize(cv.cvtColor(img,cv.COLOR_BGR2RGB), (512,1024))
plt.imshow(img4)
````

````
# Size besar n terbalik
img5 = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

 #simpan file image
 #cv.imwrite('/content/drive/MyDrive/KITTEN(2).jpg', img5)

 #tampilkan plot canvas
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.imshow(img5)
````

````
# Geometri 2D dengan openCV
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
````

````
#Menambahkan bentuk persegi
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
````

````
# Menambahkan bentuk persegi

cv.rectangle(black_img,(200,200),(300,300),(0,0,255),15)
plt.imshow(black_img)
````

````
#MENAMBAHKAN CIRCLE

cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
cv.circle(black_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(black_img)
````

````
# MENAMBAHKAN GARIS
cv.line(black_img,(0,0),(512,512),(255,255,255),5)
plt.imshow(black_img)
````

````
# MENAMBAHKAN TEKS
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
````

````
# MEMBUAT BLACK IMAGE
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
````

````
# INISIALISASI Numpy array dg type data int32
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
````

````
pts = vertices.reshape((-1,1,2))
pts
````

````
# MEMBUAT POLYLINE
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
````
